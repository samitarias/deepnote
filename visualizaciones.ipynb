```
# Start writing code here...
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns



df=pd.read_csv('cars.csv')
```

```
#Diagrama Dispersión Año vs Precio
sns.scatterplot(x='year_produced', y='price_usd',data=df)
```

```
# Mapa de calor con pivot (Ford)
ford = df[(df['manufacturer_name']=='Ford') & (df['year_produced']>2010)]
pivot_ford = ford.pivot_table(values='price_usd', index='model_name', columns='year_produced', aggfunc=np.sum)
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_ford, cmap='hot')
```

```
iris=sns.load_dataset('iris')
iris
```

```
#scatterplot
sns.scatterplot(data=iris, x= 'sepal_length',y='petal_length', hue='species')
```

```
#join plot
sns.jointplot(data=iris, x= 'sepal_length',y='petal_length',hue='species')
```

```
#box plot
sns.boxplot(x='species',y='sepal_length',data=iris)
```

```
#bar plot
sns.barplot(x='species',y='sepal_length',data=iris)
```

```
#Media de cada tipo de especie
iris.groupby('species')['sepal_length'].mean()
```

```
#Mediana
from numpy import mean,median, std
iris.groupby('species')['sepal_length'].median()
```

```
sns.barplot(x='species', y='sepal_length', data=iris, estimator=median)
```

```
#Desviación Estandar
iris.groupby('species')['sepal_length'].std()
```

```
sns.barplot(x='species', y='sepal_length', data=iris, estimator=std)
```

```
sns.pairplot(data=iris, hue='species',palette='inferno',diag_kind='kde',height=2.0)
```
